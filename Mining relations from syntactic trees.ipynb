# Mining relations from syntractic trees

## _something_ **is** _something_ (copula)

* OR: **X is Y**, where X is typically _NOUN_, _PROPN_ or _PRON_, and Y is typically _NOUN_ or _ADJ_
* For example:
    + My dog is black.
    + I am tired.
    + Gretzky is the best ice hockey player ever.
    
* Search for _"Google is X"_ structures: [Search](http://bionlp-www.utu.fi/dep_search/?db=English&search=_+%3Ensubj+Google+%3Ecop+_)
